# Load the book

```python
with open("book.txt", 'r', encoding='utf-8') as file:
    book = file.read()
```

```python
type(book)
```

# Finding chapters

## with string methods

```python
book.count("Chapter")
# not much helpful also count with in chapter
```

## regular Expressions method

```python
import re
```

```python
pattern = re.compile(r"Chapter [0-9]+")
print(pattern)
findings = re.findall(pattern, book)
print(len(findings))
```

## Finding sentences in which love is used, using regular expressions

```python
# find all words
pattern = re.compile("love")
findings = re.findall(pattern, book)
```

```python
# find with a word before and after
pattern = re.compile("[^.]*+[^a-zA-Z]love[^a-zA-Z]+[^.]*.")
findings = re.findall(pattern, book)
len(findings)
```

```python
# find with a word before and after
# pattern = re.compile("[a-zA-Z ,]* love [a-zA-Z]*")
pattern = re.compile("[^.]* love [^.]*")
findings = re.findall(pattern, book)
findings
```

# using string methods


```python
# splitting all words in a text
words = book.split() # it splits word by a space
words

# but this method is not so useful because "Octobor" and "October." are different words because . is included 
# thus we have to use regullar expressions
```

```python
# book.lower converts all uppercase to lowercase because in this case words are case insensitive
# "Word" and "word" are considers as same thing when we are counting
pattern = re.compile("[a-zA-Z]+")
findings = re.findall(pattern, book.lower())
# finding most used words
```
